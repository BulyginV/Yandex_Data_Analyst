# Изучение закономерностей, определяющих успешность игр

Интернет-магазин «Стримчик» продаёт по всему миру компьютерные игры. Из открытых источников доступны исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation). Нам нужно выявить определяющие успешность игры закономерности. Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании.  
  
У нас есть данные до 2016 года. Представим, что сейчас декабрь 2016 г., и мы планируем кампанию на 2017-й. Нужно отработать принцип работы с данными. Неважно, прогнозируем ли мы продажи на 2017 год по данным 2016-го или же 2027-й — по данным 2026 года.  
  
В наборе данных попадается аббревиатура ESRB (Entertainment Software Rating Board) — это ассоциация, определяющая возрастной рейтинг компьютерных игр. ESRB оценивает игровой контент и присваивает ему подходящую возрастную категорию, например, «Для взрослых», «Для детей младшего возраста» или «Для подростков».  

+ <code><b>«EC»</b></code> («Early childhood») — «Для детей младшего возраста»
+ <code><b>«E»</b></code> («Everyone») — «Для всех»
+ <code><b>«E10+»</b></code> («Everyone 10 and older») — «Для всех старше 10 лет»
+ <code><b>«T»</b></code> («Teen») — «Подросткам 13—19 лет»
+ <code><b>«M»</b></code> («Mature») — «От 17 лет»
+ <code><b>«AO»</b></code> («Adults Only 18+») — «Только для взрослых» 
+ <code><b>«RP»</b></code> ( «Rating Pending»») — «Рейтинг ожидается»

Напишем функцию для заполнения пропусков, если в название прописан год сезона.

```python
def fill_year_if_name(row):
    year = 0
    last_word = ''
    genre = row[3]
    for word in row[0].split():
        try:
            last_word = word
            year = int(word)
        except:
            continue
    if genre in ('Sports', 'Racing', 'Fighting'):
        if year > 2000 or last_word in ('2K6','2K8'):
            if genre == 'Sports':
                year -=1
            if genre == 'Racing':
                year -=1    
            if last_word == '2K6':
                year = 2005
            elif last_word == '2K8':
                year = 2007
            if year < 2017 and np.isnan(row[2]):
                row[2] = year
    return row

data_games = data_games.apply(fill_year_if_name, axis=1)
display(pd.concat([data_games[['year_of_release']].dtypes, data_games[['year_of_release']].isna().sum()], axis=1))
```

Напишем функцию для заполнения пропусков, если есть значение по другим платформам.

```python
not_null_values = data_games[data_games['year_of_release'].notnull()]

def fill_year_of_release(row):
    col = 'year_of_release'
    rows = not_null_values[not_null_values['name'] == row['name']]
   
    if len(rows) > 0:
        return rows.iloc[0][col]
    
    return row[col]

data_games['year_of_release'] = data_games.apply(fill_year_of_release, axis=1)
display(pd.concat([data_games[['year_of_release']].dtypes, data_games[['year_of_release']].isna().sum()], axis=1))
```

Но судя по играм и жанрам которые у них лидириуют в продажах, их возрастная аудиория 13+. Компании хотят распространить свое влияние на новое поколение.<br>

<div style="border-radius: 15px; border: 3px solid indigo; padding: 15px;">
<font color='red'>Напишем функцию <b>the_pie_func_rating</b>, которая будет строить диаграммы для сравнения влияния рейтинга ESRB на продажи по регионам.</font>
</div>

```python
def the_pie_func_rating(col1, name1, col2, name2, col3, name3):
    
    lbls = ['M', 'RP', 'E', 'E10+', 'T']
    cls = {'M': 'gold', 'RP': 'yellowgreen', 'E':'lightcoral', 
           'E10+':'lightskyblue', 'T':'dodgerblue'}
    
    fig = plt.figure(figsize=(15,5))
    ax1 = fig.add_subplot(1,3,1)
    ax2 = fig.add_subplot(1,3,2)
    ax3 = fig.add_subplot(1,3,3)

    ax1.set_title(name1)
    ax1.pie(col1, autopct='%1.1f%%', labels = col1.index, colors = [cls[key] for key in col1.index])


    ax2.set_title(name2)
    ax2.pie(col2, autopct='%1.1f%%', labels = col2.index, colors = [cls[key] for key in col2.index])


    ax3.set_title(name3)
    ax3.pie(col3, autopct='%1.1f%%', labels = col3.index, colors = [cls[key] for key in col3.index])

    plt.subplots_adjust(wspace=0.1, hspace=0.5)
```

## Шаг 5. Проверить гипотезы
* Средние пользовательские рейтинги платформ Xbox One и PC одинаковые;
* Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.  

Задать самостоятельно пороговое значение alpha.  

Пояснить:

### Первая гипотеза - Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

<div style="border-radius: 15px; border: 3px solid indigo; padding: 15px;">
<font color='purple'><b>XOne_actual_data_games</b> и <b>PC_actual_data_games</b></font> — выборки по пользовательскому рейтингу за актуальный период платформ Xbox One и PC
</div>

### Вывод по 1-ой гипотезе - Не получилось отвергнуть нулевую гипотезу

<b>ГЛАВНЫЙ ВЫВОД:</b> В целом можно сказать, что средние рейтинги платформ Xbox и PC не сильно отличаются. Гипотеза подтвердилась.



Формулировки Н0 и Н1 для 1-ой гипотезы:

<font color='blue'><b>Нулевая гипотеза: (Н0)</b></font> средний пользовательский рейтинг платформы Xbox One равен среднему пользовательскому рейтингу платформы PC    

<font color='red'><b>Альтернативная гипотеза: (Н1)</b></font> средний пользовательский рейтинг платформы Xbox One отличается от среднего пользовательского рейтинга платформы PC.  

<b>Для проверки гипотезы</b> используем гипотезу о равенстве средних двух генеральных совокупностей поскольку данные были собраны в одном актуальном периоде (2013-2016) и их характеристики схожые.

Пороговое значение alpha (критический уровень статистической значимости) зададим равным 3%, т.к. в датасете присутствет большое количество пропусков.

```python
alpha = 0.03
```

```python
# XOne_actual_data_games = actual_data_games.query('platform =="XOne"')
PC_actual_data_games = actual_data_games.query('platform =="PC"')

results = st.ttest_ind(
    XOne_actual_data_games['user_score'],
    PC_actual_data_games['user_score'],
    equal_var=False,  
    nan_policy='omit' 
)

print('Среднее XOne:', round(XOne_actual_data_games['user_score'].mean(), 2))
print('Среднее PC:  ', round(PC_actual_data_games['user_score'].mean(), 2))

print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

### Вторая гипотеза - Средние пользовательские рейтинги жанров Action и Sports разные.

<div style="border-radius: 15px; border: 3px solid indigo; padding: 15px;">
<font color='purple'><b>action_actual_data_games</b> и <b>sports_actual_data_games</b></font> — выборки по пользовательскому рейтингу за актуальный период по жанрам Action и Sports
</div>

### Вывод по 2-ой гипотезе - Отвергаем нулевую гипотезу

<b>ГЛАВНЫЙ ВЫВОД:</b> Рейтинги жанров Action и Sports различаются. Гипотеза подтвердилась.

Формулировки Н0 и Н1 для 2-ой гипотезы:

<font color='blue'><b>Нулевая гипотеза: (Н0)</b></font> средний пользовательский рейтинг жанра Action равен среднему пользовательскому рейтингу жанра Sports.    

<font color='red'><b>Альтернативная гипотеза: (Н1)</b></font> средний пользовательский рейтинг жанра Action отличается от среднего пользовательского рейтинга жанра Sports.  

<b>Для проверки гипотезы</b> используем гипотезу о равенстве средних двух генеральных совокупностей поскольку данные были собраны в одном актуальном периоде (2013-2016) и их характеристики схожые.

Пороговое значение alpha (критический уровень статистической значимости) зададим равным 3%, т.к. в датасете присутствет большое количество пропусков.

```python
alpha = 0.03
```

```python
action_actual_data_games = actual_data_games.query('genre =="Action"')
sports_actual_data_games = actual_data_games.query('genre =="Sports"')

results = st.ttest_ind(
    action_actual_data_games['user_score'],
    sports_actual_data_games['user_score'],
    nan_policy='omit' 
)

print('Среднее Action:', round(action_actual_data_games['user_score'].mean(), 2))
print('Среднее PC:    ', round(sports_actual_data_games['user_score'].mean(), 2))

print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```
