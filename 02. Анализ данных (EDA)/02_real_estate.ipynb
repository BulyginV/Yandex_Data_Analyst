### 3.3. Посчитать и добавить в таблицу этаж квартиры (варианты — первый, последний, другой)

Для выполнения задания, избавимся от оставшихся пропусков столбца <code><b>floors_total</b></code> (всего этажей в доме).

```python
data_project_2['floors_total'] = data_project_2['floors_total'].fillna(0)

def floor_total_nan(row):
   
    if row['floors_total'] == 0:
        return (row['floor'] + 1)
    else:
        return row['floors_total']
       
data_project_2.loc[:, 'floors_total'] = data_project_2.apply(floor_total_nan, axis=1)
display(pd.concat([data_project_2[['floors_total']].dtypes, data_project_2[['floors_total']].isna().sum()], axis=1))
```

Создадим функцию категоризации этажей.

```python
def floor_cat(row):
    if row['floor'] == 1:
        return 'Первый этаж'
    if row['floor'] == row['floors_total']:
        return 'Последний этаж'
    else:
        return 'Другой этаж'    
```

Просчитаем и добавим новый столбец <code><b>data_project_2['floor_type']</b></code> (тип этажа квартиры)

<div style="border-radius: 15px; border: 3px solid indigo; padding: 15px;">

</div>

```python
data_project_2['floor_type'] = data_project_2.apply(floor_cat, axis=1)
display(data_project_2.head())
```
