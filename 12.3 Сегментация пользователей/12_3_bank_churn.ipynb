### 5.1. Проверка гипотезы о равенстве доходов между клиентами с одним и двумя продуктами

<b>Проверим гипотезу:</b><br>
H0: Различие дохода у клиентов с 1 продуктом и у клиентов с 2 продуктами статистически незначимо.<br>
H1: Различие дохода у клиентов с 1 продуктом и у клиентов с 2 продуктами статистически значимо.<br>

```python
alpha = 0.05
```

<div style="border-radius: 15px; border: 3px solid indigo; padding: 15px;">
<code><b>product_one_salary</b></code> — выборка по зарплатам клиентов с одним продуктом
</div>

```python
product_one_salary = bank_dataset.loc[bank_dataset['products']==1, 'salary']
display(product_one_salary.head())
```

<div style="border-radius: 15px; border: 3px solid indigo; padding: 15px;">
<code><b>product_two_salary</b></code> — выборка по зарплатам клиентов с двумя продуктами
</div>

```python
product_two_salary = bank_dataset.loc[bank_dataset['products']==2, 'salary']
display(product_two_salary.head())
```

В исследовательском анализа данных мы убедились, что распределение заработной платы далеко от нормального, а количество наблюдений достаточно (более 5000), для расчёта <code><b>p_value</b></code> используем критерий Манна-Уитни.

```python
result = st.mannwhitneyu(product_one_salary, product_two_salary, True, 'two-sided')
print('p-значение: ', result.pvalue)

if (result.pvalue < alpha):
    print("Отвергаем нулевую гипотезу: разница статистически значима")
else:
    print("Не получилось отвергнуть нулевую гипотезу, вывод о различии сделать нельзя") 
```

Следовательно, доходы клиентов с одним и двумя продуктами не отличаются.

### 5.2. Проверка гипотезы о равенстве оттока клиентов между сегментами

<b>Проверим гипотезу:</b><br>
H0: Различие оттока в сегментах 1 и 2 статистически незначимо.<br>
H1: Различие оттока в сегментах 1 и 2 статистически значимо.<br>

<div style="border-radius: 15px; border: 3px solid indigo; padding: 15px;">
<code><b>segment_one_churn</b></code> — выборка по оттоку клиентов первого сегмента
</div>

```python
segment_one_churn = bank_dataset.loc[bank_dataset['segment_products']==1, 'churn']
display(segment_one_churn.head())
```

<div style="border-radius: 15px; border: 3px solid indigo; padding: 15px;">
<code><b>segment_two_churn</b></code> — выборка по оттоку клиентов второго сегмента
</div>

```python
segment_two_churn = bank_dataset.loc[bank_dataset['segment_products']==2, 'churn']
display(segment_two_churn.head())
```

При проверке гипотез мы имеем дело с бинарными данными, поэтому применим ztest. Количество наблюдений достаточно.

```python
from statsmodels.stats import weightstats as stests
ztest ,pval = stests.ztest(segment_one_churn, x2=segment_two_churn, value=0,alternative='two-sided')
print('p-значение: ', float(pval))

if (pval < alpha):
    print("Отвергаем нулевую гипотезу: разница статистически значима")
else:
    print("Не получилось отвергнуть нулевую гипотезу, вывод о различии сделать нельзя")
```

Можно сделать вывод о том, что разница в доле оттока между сегментами статистически значима.

### 6.2. Формирование рекомендаций по результатам исследования

<div style="border-radius: 15px; border: 3px solid green; padding: 15px;">

- Главный вывод исследования это давать не больше и не меньше двух продуктов в одни руки и тому есть логичное объяснение. Клиенты с двумя продуктами в 4 раза реже принимают решение уйти из банка чем клиенты с другим количеством продуктов.<br>
    
- Далее разберемся что это должны быть за продукты. Это должна быть продуктовая корзина 50/50 т.е. один дебетовый продукт и один кредитный продукт и только в такой пропорции. И в этом нам поможет выверенная стратегия кросс-продаж.<br>
    
- Рассмотрим более детально стратегия кросс-продаж продуктов:<br>
    - клиент оформил кредитный продукт (ипотека/наличные) или кредитную карту, в таком случае ему необходимо предложить (допродать) выгодную карту с повышенным кэшбэком и возможно нулевым обслуживанием;<br>
    - клиент оформил себе дебетовый продукт (депозит, дебетовая карта, инвестиционный счет), в таком случае ему необходимо предложить (допродать) выгодную кредитную карту с повышенным льготным периодом, рефинансирование ипотеки (если она у него уже есть в другом банке), рефинансирование кредита на наличные (если такой кредит есть в другом банке).<br>
    
- Теперь проработаем параметр "где", а именно продажи по территориальному признаку. Стратегию необходимо применить во всех городах присутствия, соответственно обновить продуктовые линейки, а точнее синхронизировать их.<br>
    
- Далее определимся с параметром "кому" мы продаём, кто наш целевой клиент. Наш целевой клиент это преимущественно мужчина в возрасте от 18 до 45 лет т.к. по полученным данным из нашего исследования эта категория меньше всего подвержена оттоку.<br>
    
- И наконец "как" мы можем повлиять на привлечение наших целевых клиентов:<br>
    - через повышение уровня одобрения кредитных продуктов путём частичного пересмотра скоринговой модели, а точнее начисления скоринговых баллов за определенные признаки. Такие как, повысить скоринговый балл мужчинам в возрасте от 18 до 45, с доходом до 100000 руб;<br>
    - через маркетинговые активности (реклама в интернете рассчитанная преимущественно на мужскую аудиторию, реклама телевизору, всевозможные "хайпы" и рекламные интеграции в юмористических и не только программах, каналах). Реклама в соц. сетях и мессенджерах.<br>

- И отдельно нужно определенно менять метод заполнения значений по признаку "зарплата". Это очень важный параметр при работе нашей скоринговой модели. Здесь следует максимально закупиться базами, такими как база БКИ (Бюро кредитных историй) и не только. Возможно разработать сценарий запроса 2НДФЛ с минимальным участием клиента и точно не на этапе оформления кредитной заявки. (Это подойдет для категории клиентов, которые изначально оформляют себе дебетовый продукт и далее уже на этапе кросс-продажи нужно реализовать аккуратный сценарий запроса корректных цифр по признаку зарплата).<br>

- Конечно для более эффективной проработки бизнесовой составляющей нам нужно понимать основные KPI банка и если на данный момент его основными KPI является увеличение комиссионного дохода, смещение акцентов от клиентов физ. лиц к клиентам юр. лицам, соответственно у нас поменяется продуктовая логика и возможно показатель оттока уйдет на второй план, а допустим показатель NPV выйдет на первый. У любого банка есть свои акционеры и нужно понимать его приоритетные задачи на 1/2/5 лет.<br>
